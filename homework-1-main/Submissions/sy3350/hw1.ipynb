## Project Overview <a class="anchor" id="overview"></a>

AI Basketball Analysis is an **AI-powered** basketball video analysis project that focuses on **object detection** and **pose estimation** to analyze shots and shooting form. 

This project highlights **object detection** and *human pose estimation* techniques.


### What the app can do
- Upload a basketball video/image for analysis (web app)
- Send an image to a REST endpoint and receive JSON results (API)
- Estimate shooting pose metrics using *OpenPose* (body keypoints like elbow/knee angles) 


## How It Works <a class="anchor" id="how"></a>

The project uses object detection models to identify the basketball and shot events, and OpenPose to compute human body keypoints such as elbow and knee angles.

These keypoints are used to analyze shooting mechanics and visualize successful and missed shots with different colors.

The application can be run locally or accessed through an API endpoint.


## How It Works (High Level) <a class="anchor" id="how"></a>

The repository describes a pipeline that:
1) detects relevant objects (e.g., ball/hoop) and shot outcomes,
2) runs pose estimation with OpenPose to compute body keypoints,
3) reports shot + pose results through the web UI and an API endpoint like `POST /detection_json`.
