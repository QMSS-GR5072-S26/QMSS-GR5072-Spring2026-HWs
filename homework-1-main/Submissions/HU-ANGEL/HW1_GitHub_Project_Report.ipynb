This GitHub project analyzes the classic *Iris Dataset*, which contains measurements of flower characteristics such as sepal length, sepal width, petal length, and petal width.

The goal of this project is to explore relationships between flower features and visualize how different species cluster together.

This project is interesting because it demonstrates basic **data science workflows**, including loading data, visualization, and exploratory analysis using Python.

*I chose this repository because it provides a clear and beginner-friendly example of how GitHub can be used to share data analysis projects.*


## 2. Dataset Description <a class="anchor" id="bullet2"></a>

### Iris Flower Attributes

The dataset contains:

- Sepal length
- Sepal width
- Petal length
- Petal width
- Species label

These variables help classify flowers into three species.

Two important points:

- **Petal length is highly predictive**
- *Sepal width shows more overlap*

This makes the dataset ideal for learning classification and visualization techniques.
