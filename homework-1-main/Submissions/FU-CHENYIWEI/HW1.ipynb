# Project Overview <a class="anchor" id="section1"></a>

The project I selected is **Customer Churn Prediction in the Banking Industry**.

GitHub Repository Link:
https://github.com/chenyiweifu108/customer_churn_prediction.github.io

This project applies supervised machine learning models to predict whether a bank customer is likely to leave (churn).

__Customer retention is significantly more cost-effective than customer acquisition.__

By predicting churn early, financial institutions can reduce revenue loss and improve long-term profitability.


# Modeling Approach <a class="anchor" id="section2"></a>

The project implements several supervised machine learning models:

- Logistic Regression
- Random Forest
- K-Nearest Neighbors (KNN)
- Feature Importance Analysis
- Cross-validation

## Why This Project Is Important <a class="anchor" id="section2a"></a>

*Churn prediction directly impacts business strategy.*

By identifying high-risk customers, banks can:

- Launch targeted retention campaigns
- Adjust pricing and service strategies
- Optimize long-term customer lifetime value

# Key Results <a class="anchor" id="section4"></a>

The Random Forest model achieved the strongest performance:

- Accuracy: 85.9%
- Precision: 80.4%
- Recall: 41.1%
- AUC: 0.85

**Random Forest outperformed Logistic Regression and KNN.**

Feature importance analysis revealed that the most influential predictors were:

- Credit Score
- Age
- Account Balance
- Geography
